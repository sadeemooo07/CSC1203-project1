```python
# Parent Class (Person)
class Person:
    def __init__(self, name, student_id):
        self.name = name # Public attribute
        self.__student_id = student_id# Private attribute
   
    def get_id(self): # Getter method للحصول على قيمة الـ ID
        return self.__student_id
    
    def set_id(self, new_id):# Setter method لتعديل قيمة الـ ID
        self.__student_id = new_id

# Child Class (Student)
class Student(Person):
    def __init__(self, name, student_id, grade):# استدعاء constructor من الكلاس الأب
         super().__init__(name, student_id)
         self.grade = grade

    def display(self): # عرض بيانات الطالب
        print(f"Name: {self.name}, ID: {self.get_id()}, Grade: {self.grade}")

# File Handling
FILE_NAME = "data.txt"# اسم الملف اللي نخزن فيه البيانات
def load_students():# تحميل البيانات من الملف عند تشغيل البرنامج
    students = []  # قائمة لتخزين الطلاب
    
    try:
        with open(FILE_NAME, "r") as file:
            for line in file:
                # تقسيم البيانات (name, id, grade)
                name, student_id, grade = line.strip().split(",")
                
                # إنشاء كائن طالب وإضافته للقائمة
                students.append(Student(name, student_id, int(grade)))
    
    # إذا الملف غير موجود، نتجاهل الخطأ
    except FileNotFoundError:
        pass
    
    return students


# حفظ البيانات في الملف
def save_students(students):
    with open(FILE_NAME, "w") as file:
        for s in students:
            # كتابة كل طالب في سطر
            file.write(f"{s.name},{s.get_id()},{s.grade}\n")

# Operations (CRUD)

# إضافة طالب
def add_student(students):
    name = input("Enter name: ")
    student_id = input("Enter ID: ")

    # إدخال الدرجة مع التحقق
    grade = int(input("Enter grade (0-100): "))
    
    if grade < 0 or grade > 100:
        print("Invalid grade!")
        return

    # إنشاء طالب جديد
    students.append(Student(name, student_id, grade))
    print("Student added successfully!")


# حذف طالب
def delete_student(students):
    student_id = input("Enter ID to delete: ")
    
    for s in students:
        if s.get_id() == student_id:
            students.remove(s)
            print("Deleted successfully!")
            return
    
    print("Student not found!")


# تعديل بيانات طالب
def update_student(students):
    student_id = input("Enter ID to update: ")
    
    for s in students:
        if s.get_id() == student_id:
            s.name = input("New name: ")
            s.grade = int(input("New grade: "))
            print("Updated successfully!")
            return
    
    print("Student not found!")


# عرض جميع الطلاب
def display_students(students):
    if not students:
        print("No students found.")
    
    for s in students:
        s.display()



# Main Program (Menu)

def main():
    # تحميل البيانات من الملف
    students = load_students()

    # حلقة مستمرة لعرض القائمة
    while True:
        print("\n--- Student Management System ---")
        print("1. Add Student")
        print("2. Delete Student")
        print("3. Update Student")
        print("4. Display Students")
        print("5. Exit")

        choice = input("Choose: ")

        # تنفيذ الخيار حسب اختيار المستخدم
        if choice == "1":
            add_student(students)

        elif choice == "2":
            delete_student(students)

        elif choice == "3":
            update_student(students)

        elif choice == "4":
            display_students(students)

        elif choice == "5":
            # حفظ البيانات قبل الخروج
            save_students(students)
            print("Data saved. Goodbye!")
            break

        else:
            print("Invalid choice!")

# تشغيل البرنامج

if __name__ == "__main__":
    main()            
```
