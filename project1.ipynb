```python
# تعريف كلاس يمثل شخص (Parent Class)
class Person:
    
    # دالة البناء (constructor) لتخزين الاسم والآيدي
    def __init__(self, name, student_id):
        self.name = name  # متغير عام لتخزين الاسم
        self.__student_id = student_id  # متغير خاص لتخزين الآيدي

    # دالة لإرجاع الآيدي (Getter)
    def get_id(self):
        return self.__student_id

    # دالة لتعديل الآيدي (Setter)
    def set_id(self, new_id):
        self.__student_id = new_id


# تعريف كلاس الطالب ويرث من Person
class Student(Person):

    # دالة البناء للطالب
    def __init__(self, name, student_id, grade):
        super().__init__(name, student_id)  # استدعاء كلاس الأب
        self.grade = grade  # تخزين الدرجة

    # دالة لعرض بيانات الطالب
    def display(self):
        print(f"Name: {self.name}, ID: {self.get_id()}, Grade: {self.grade}")


# اسم الملف المستخدم لتخزين البيانات
FILE_NAME = "data.txt"


# دالة لتحميل الطلاب من الملف
def load_students():
    students = []  # قائمة فارغة لتخزين الطلاب
    try:
        with open(FILE_NAME, "r") as file:  # فتح الملف للقراءة
            for line in file:  # قراءة كل سطر
                name, student_id, grade = line.strip().split(",")  # تقسيم البيانات
                students.append(Student(name, student_id, int(grade)))  # إضافة الطالب
    except FileNotFoundError:  # في حال الملف غير موجود
        pass
    return students  # إرجاع القائمة


# دالة لحفظ البيانات في الملف
def save_students(students):
    with open(FILE_NAME, "w") as file:  # فتح الملف للكتابة
        for s in students:
            file.write(f"{s.name},{s.get_id()},{s.grade}\n")  # كتابة البيانات


# دالة لإضافة طالب
def add_student(students):
    name = input("Enter name: ")  # إدخال الاسم
    student_id = input("Enter ID: ")  # إدخال الآيدي

    grade = int(input("Enter grade (0-100): "))  # إدخال الدرجة
    
    if grade < 0 or grade > 100:  # التحقق من صحة الدرجة
        print("Invalid grade!")
        return

    students.append(Student(name, student_id, grade))  # إضافة الطالب
    print("Student added successfully!")


# دالة لحذف طالب
def delete_student(students):
    student_id = input("Enter ID to delete: ")  # إدخال الآيدي

    for s in students:
        if s.get_id() == student_id:  # البحث عن الطالب
            students.remove(s)  # حذف الطالب
            print("Deleted successfully!")
            return

    print("Student not found!")  # في حال عدم وجوده


# دالة لتعديل بيانات طالب
def update_student(students):
    student_id = input("Enter ID to update: ")  # إدخال الآيدي

    for s in students:
        if s.get_id() == student_id:
            s.name = input("New name: ")  # تعديل الاسم
            s.grade = int(input("New grade: "))  # تعديل الدرجة
            print("Updated successfully!")
            return

    print("Student not found!")


# دالة لعرض جميع الطلاب
def display_students(students):
    if not students:
        print("No students found.")

    for s in students:
        s.display()  # عرض كل طالب


# الدالة الرئيسية
def main():
    students = load_students()  # تحميل البيانات

    while True:  # حلقة مستمرة
        print("\n--- Student Management System ---")
        print("1. Add Student")
        print("2. Delete Student")
        print("3. Update Student")
        print("4. Display Students")
        print("5. Exit")
        print("6. Clear All Data")

        choice = input("Choose: ")  # اختيار المستخدم

        if choice == "1":
            add_student(students)

        elif choice == "2":
            delete_student(students)

        elif choice == "3":
            update_student(students)

        elif choice == "4":
            display_students(students)

        elif choice == "5":
            save_students(students)  # حفظ البيانات
            print("Data saved. Goodbye!")
            break
        elif choice == "6":
            students.clear()  # يمسح القائمة
            open("data.txt", "w").close()  # يفرغ الملف
            print("All data cleared!")

        else:
            print("Invalid choice!")


# تشغيل البرنامج
if __name__ == "__main__":
    main()
```
